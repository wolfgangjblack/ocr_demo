```python
import cv2
import pytesseract
from PIL import Image
```

```python
def extract_text_from_image(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Try different preprocessing techniques
    
    # 1. Original grayscale
    text1 = pytesseract.image_to_string(gray)
    
    # 2. Binary threshold
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    text2 = pytesseract.image_to_string(binary)
    
    # 3. Adaptive threshold
    adaptive = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    text3 = pytesseract.image_to_string(adaptive)
    
    # 4. Otsu's threshold
    _, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    text4 = pytesseract.image_to_string(otsu)
    
    return f"Grayscale:\n{text1}\n\nBinary:\n{text2}\n\nAdaptive:\n{text3}\n\nOtsu:\n{text4}"
```
